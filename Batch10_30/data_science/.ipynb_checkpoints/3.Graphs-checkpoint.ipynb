```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib inline
x = np.arange(1,11)
y = np.random.random(10)
z = np.random.random(10)
p = np.random.randint(20,100,10)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x,y,c=z,s=p)
ax.set_xlim([-5,15])
ax.set_ylim([-1,2])
plt.show()
```

```python
import psutil
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import time
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111)
y = [ psutil.cpu_percent() for var in range(50) ]
x = [ var for var in range(50)]
ax.plot(x,y)
def change(i):
    global x,y
    x.pop(0)
    y.pop(0)
    x.append(x[-1]+1)
    y.append(psutil.cpu_percent())
    ax.clear()
    ax.plot(x,y,'r^--')
ani = FuncAnimation(fig,change,2000)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
from time import sleep
from psutil import cpu_percent as cpu
plt.rcParams["animation.html"] = 'jshtml'
%matplotlib notebook
x = [ var for var in range(50)]
y = [ cpu() for var in range(50) ]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y,'go--')
plt.show()
```

```python
while True : 
    x.pop(0)
    y.pop(0)
    x.append(x[-1]+1)
    y.append(cpu())
    ax.clear()
    ax.plot(x,y,'go--')
    fig.canvas.draw()
    sleep(1)
```

```python
for var in range(10): 
    print(psutil.cpu_percent())
    time.sleep(2)
```
